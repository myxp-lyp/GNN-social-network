```python
import csv

csv.field_size_limit(500 * 1024 * 1024)
f = csv.reader(open('user_new.csv','r',encoding='utf-8',errors='ignore'))

Mmap={}
```

```python
f.iloc[0][9]
```

```python
for i in range(len(f)):
    if f.iloc[i][9]=='f':
        f.loc[i:i,('gender')]=0
    elif f.iloc[i][9]=='m':
        f.loc[i:i,('gender')]=1
```

```python
for i in f:
    Mmap[i[1]]=i[0]
```

```python
Mmap.keys()
```

```python
f = open('idlist.txt','w') 
for i in Mmap.keys():
    f.write(i)
    f.write('\n')
f.close()
```

```python
f = open('total_f.txt','r')
a = f.readlines()  
f = open('total_idmap.txt','w') 
j=1
for i in a:
    if j%2:
        ii=i.split()
        f.write(Mmap[ii[0]])
        f.write('\n')
    else:
        for ii in i.split():
            f.write(Mmap[ii])
            f.write(' ')
        f.write('\n')
    j=j+1
f.close()
```

```python
f = open('Gmerge.txt','r')
a = f.readlines()
f = open('G_idmap.txt','w') 
j=1
for i in a:
    if len(i.split()):
        x,y=i.split()
        if str(x) in Mmap.keys() and str(y) in Mmap.keys():
            f.write(Mmap[str(x)]+' '+Mmap[str(y)])
            f.write('\n')
f.close()
```
